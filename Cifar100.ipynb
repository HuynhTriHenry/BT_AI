```python
from keras.datasets import cifar100
(x_train,y_train),(x_test,y_test)= cifar100.load_data()
```

```python
import matplotlib.pyplot as plt
hinh= x_train[10]   
plt.imshow(hinh,cmap=plt.cm.binary)
plt.show
print("So nay la:",y_train[10])
```

```python
for i in range(30):
  plt.subplot(6,8,i+1)  
  plt.tight_layout()
  plt.imshow(x_train[i],cmap='gray',interpolation='none')
  plt.title("So:{}".format(y_train[i]))
  plt.xticks([])
  plt.yticks([])
```

```python
from keras.models import Sequential
from keras.layers import Dense

model= Sequential()
model.add(Dense(512, input_shape = (32*32*3,), activation = 'relu'))
model.add(Dense(100,activation= 'softmax'))
model.summary()
```

```python
x_train = x_train.reshape((50000,32*32*3,))
x_train = x_train.astype('float32')/255
x_test = x_test.reshape((10000,32*32*3,))
x_test = x_test.astype('float32')/255
```

```python
from keras.utils import to_categorical #Lệnh chuyển qua 

y_train = to_categorical(y_train, 100) #chuyển sang 10 kênh
y_test = to_categorical(y_test, 100)
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs = 800, batch_size= 128) # Mỗi lần học chỉ 128 dữ liệu, không học hết 60000 dữ liệu , học tổng cộng 10 lầnlần
test_loss, test_acc = model.evaluate(x_test,y_test)
print('test_accuracy:',test_acc)
print('test_loss:',test_loss)
```
